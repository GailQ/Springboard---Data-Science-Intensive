```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.preprocessing import StandardScaler

data = dfvegan[["rating", "fat"]]
data = data.fillna(value=0)
dfvegan.head(10)
print(data)
data = data.as_matrix().astype("float64", copy = False)
print(data)
stscaler = StandardScaler().fit(data)
data = stscaler.transform(data)
print(data)
#plt.scatter(data[:,0], data[:,1])
plt.xlabel("Rating")
plt.ylabel("Fat")
plt.title("Rating - Fat")

dbsc = DBSCAN(eps = .5, min_samples = 15).fit(data)

labels = dbsc.labels_
core_samples = np.zeros_like(labels, dtype = bool)
core_samples[dbsc.core_sample_indices_] = True

unique_labels = np.unique(labels)
colors = plt.cm.Spectral(np.linspace(0,1, len(unique_labels)))
 
for (label, color) in zip(unique_labels, colors):
    class_member_mask = (labels == label)
    xy = data[class_member_mask & core_samples]
    plt.plot(xy[:,0],xy[:,1], 'o', markerfacecolor = color, markersize = 10)
    
    xy2 = data[class_member_mask & ~core_samples]
    plt.plot(xy2[:,0],xy2[:,1], 'o', markerfacecolor = color, markersize = 5)
plt.title("DBSCAN on Rating and Fat")
plt.xlabel("Rating")
plt.ylabel("Fat")
```

```python
from collections import defaultdict
features = dfvegan.columns
total = defaultdict(int)
for i in features:
    try:                 # avoid errors when values are not of type(int)
        count = dfvegan[dfvegan[i] > 0].copy()
        total[i] = len(count)
    except:
        pass
feature_counts = list(total.values())
feature_counts.sort(reverse=True)
for fc in feature_counts[:372]:      
    for key, t in total.items():
        if t == fc:
            print ("{:<20}{:,}".format(key, t))         
```
